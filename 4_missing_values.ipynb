### isnull() and notnull()

```python
df=pd.read_csv("C:\\Users\\nikhil\\Desktop\\export.csv")
df
```

```python
df.isnull()
# all are false it means there are no missing value in the data.
```

```python
# adding some missing values into the data
df=pd.read_csv("C:\\Users\\nikhil\\Desktop\\export.csv")
df
```

```python
df.isnull()
```

```python
# count null values column wise
# here "GROWTH" column has 4 na values
df.isnull().sum()
```

```python
# to count total no of null values into the data
df.isnull().sum().sum()
```

```python
# notnull() function is exact opposite of isnull()
df=pd.read_csv("C:\\Users\\nikhil\\Desktop\\export.csv")
df
```

```python
df.notnull()
```

```python
# how many correct values are there in the data column wise
df.notnull().sum()
```

```python
# total correct values into the data
df.notnull().sum().sum()
```
