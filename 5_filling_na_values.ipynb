### Filling Na values:

### fillna()

```python
import pandas as pd
```

```python
df=pd.read_csv("C:\\Users\\nikhil\\Desktop\\export.csv")
df
```

```python
"""
Signature:
df.fillna(
    value=None,
    method=None,
    axis=None,
    inplace=False,
    limit=None,
    downcast=None,
    **kwargs,
)

"""
```

```python
# all na value fill by value 0
df.fillna(0)
```

```python
# all na value fill by value 2 
df.fillna(2)
```

```python
# to fill with particular value in particular column,we pass the dictionary.
df.fillna({"NAME":"MAHARAJA","COMPANIES":"SALES"})
```

```python
df=pd.read_csv("C:\\Users\\nikhil\\Desktop\\export.csv")
df
```

```python
#method parameter:fill na/NaN values using specified method
#method : {'backfill', 'bfill', 'pad', 'ffill', None}
#pad / ffill: propagate last valid observation forward to next valid
#backfill / bfill: use NEXT valid observation to fill gap
df.fillna(method="ffill")
```

```python
df=pd.read_csv("C:\\Users\\nikhil\\Desktop\\export.csv")
df
```

```python
df.fillna(method="bfill")
```

```python
# axis parameter:
# used with method parameter ,without method we cant use axis parameter
df=pd.read_csv("C:\\Users\\nikhil\\Desktop\\export.csv")
df
```

```python
df.fillna(method="ffill",axis=0)
```

```python
df.fillna(method="ffill",axis=1)
```

```python
# limit parameter:
# value parameter/method parameter is necessary to use limit parameter
df=pd.read_csv("C:\\Users\\nikhil\\Desktop\\export.csv")
df
```

```python
# fill na  with 0 ,only 1 time
df.fillna(0,limit=1)
```

```python
df.fillna(0,limit=2)
```

```python
#inplace parameter:
df=pd.read_csv("C:\\Users\\nikhil\\Desktop\\export.csv")
df
```

```python
df.fillna(5,inplace=True)
df
```
