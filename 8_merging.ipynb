### Pandas Merging:

```python
import pandas as pd
```

```python
df1=pd.DataFrame({"ID":[1,2,3,4],"Class":[9,10,11,12]})
df1
```

```python
df2=pd.DataFrame({"ID":[1,2,3,5],"Name":['A','B','C','D']})
df2
```

```python
''' 
pd.merge(
    left,
    right,
    how='inner',
    on=None,
    left_on=None,
    right_on=None,
    left_index=False,
    right_index=False,
    sort=False,
    suffixes=('_x', '_y'),
    copy=True,
    indicator=False,
    validate=None,
)'''
```

```python
pd.merge(df1,df2)
```

```python
# on parameter:
pd.merge(df1,df2,on="ID")
```

```python
pd.merge(df2,df1)

# here df2 columns come first than df1
```

```python
pd.merge(df1,df2)
```

```python
# how parameter:
pd.merge(df1,df2,how="inner")
```

```python
pd.merge(df1,df2,how="left")
```

```python
pd.merge(df1,df2,how="right")
```

```python
pd.merge(df1,df2,how="outer")
```

```python
# indicator parameter:
pd.merge(df1,df2,how="left",indicator=True)
```

```python
pd.merge(df1,df2,how="inner",indicator=True)
```

```python
pd.merge(df1,df2,how="outer",indicator=True)
```

```python
# left_index and right_index parameter:
df3=pd.DataFrame({"ID":[1,2,3,4],"Class":[7,8,9,10]})
df3
```

```python
df4=pd.DataFrame({"ID":[5,6,7,8],"Name":['A','B','C','D']})
df4
```

```python
pd.merge(df3,df4)
```

```python
pd.merge(df3,df4,left_index=True,right_index=True)
```

```python
df3
```

```python
df5=pd.DataFrame({"ID":[1,2,3,4],"Class":[7,8,9,10]})
df5
```

```python
pd.merge(df3,df5)
```

```python
pd.merge(df3,df5,left_index=True,right_index=True)
```

```python
# suffixes parameter:
pd.merge(df3,df5,suffixes=('_11th','_12th'))
```

```python
pd.merge(df3,df5,on="ID",suffixes=('_11th','_12th'))
```

```python
pd.merge(df3,df5,on="Class",suffixes=('_11th','_12th'))
```
